```python
import numpy as np
```

```python
a = np.mod(5,3) # 取模运算
print(a)
a = 5%3
print(a)
```

```python
x = np.linspace(0,10,11)
print(x)
```

```python
x = np.linspace(0,2*np.pi,int(2*np.pi/0.01))
y = np.cos(x)
import matplotlib.pyplot as plt
plt.plot(x,y,label='y=cos(x)')
y = np.sin(x)
plt.plot(x,y,label='y=sin(x)')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111,projection='3d')
n = 100
f = lambda min_val,max_val,num:min_val + (max_val-min_val)*np.random.randn(num)
x = f(10,40,n)
y = f(-100,100,n)
z = f(50,200,n)
ax.scatter(x,y,z,marker='o',color='r')
ax.plot(x,y,z,marker='o',color='b',markersize=3)
ax.set_xlim(10,40)
ax.set_ylim(-100,100)
ax.set_zlim(50,200)
# ax.set_aspect('equal')
plt.show()
```
