一个**神经网络**就是一个模仿人类大脑激发学习过程的框架。神经网络被用于从数据中心识别隐藏的模式。    
正如所有的学习算法，神经网络处理的是数字。因此，如果想要实现处理现实世界中任何包含图像、文字、传感器等的任务，就必须转换成数值的形式，然后将其输入到一个神经网络。     

神经网络由一层层神经元组成。这些神经元模拟人类大脑中的生物神经元。每一层都是一组独立的神经元，这些神经元与相邻层的神经元相连。
输入层对应输入数据，输出层对应输出结果，输入层和输出层之间的层统称为隐藏层。    

而深度神经网络就是由多个隐藏层组成的神经网络。

**感知器**是一个独立的神经元，它负责执行所有的计算。   
该神经元将多个输入用不同的权重系数融合起来，并加上偏差值来计算输出。   
这是一个简单的线性方程，它将输入值和感知器的输出值关联起来

```python
import numpy as np
import neurolab as nl
import matplotlib.pyplot as plt

# 定义一些输入数据及其对应标签
data = np.array([[0.3, 0.2], [0.1, 0.4], [0.4, 0.6], [0.9, 0.5]])
labels = np.array([[0], [0], [0], [1]])
# 将这些点画出
plt.figure()
plt.scatter(data[:,0], data[:,1])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Input data')
# 定义一个感知器perceptron，它有两个输入。
# 该函数还需要限定输入数据的最大值和最小值，感知器第一个参数的每个元素中指定参数的最大值和最小值
perceptron = nl.net.newp([[0, 1],[0, 1]], 1)
# 训练感知器
# epochs--指定了训练数据集需要完成的测试次数；show-指定了显示训练过程的频率
# lr-感知器的学习速度，学习速度是指学习算法在参数空间中的搜索步长
error = perceptron.train(data, labels, epochs=50, show=15, lr=0.01)
# 画出结果
plt.figure()
plt.plot(error)
plt.xlabel('Number of epochs')
plt.ylabel('Training error')
plt.grid()
plt.title('Training error progress')

plt.show()
```
